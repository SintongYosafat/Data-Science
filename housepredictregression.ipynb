```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df_Train = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/train.csv") 
df_Test = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/test.csv")
```

```python
df_Train = pd.get_dummies(df_Train)
df_Test = pd.get_dummies(df_Test)
```

```python
df_Test["SalePrice"] = df_Train["SalePrice"]
```

```python
df_Train.dropna(axis=1, inplace=True)
df_Test.dropna(axis=1, inplace=True)
```

```python
set1= set(df_Train.columns)
set2= set(df_Test.columns)

intersection = set1.intersection(set2)

df1 = df_Train[intersection]
df2 = df_Test[intersection]
```

```python
X_train = df1.drop(["SalePrice"], axis = 1)
y_train = df1["SalePrice"]
X_test = df2.drop(["SalePrice"], axis = 1)
cid = df2["Id"]
```

```python
rf = RandomForestRegressor()
model_rf = rf.fit(X_train,y_train)
rf_predict = model_rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
pd.options.display.max_rows = None
results = pd.DataFrame({"id" : cid,
                        "SalePrice" : rf_predict})

results
```
